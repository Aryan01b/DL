```python
import numpy as np
import pandas as pd
from tensorflow.keras.datasets import boston_housing
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error
import time
```

```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
x_train.shape
```

```python
start_time = time.time()

model = Sequential()
model.add(Dense(1, input_dim=x_train.shape[1], activation='linear'))  # Single output neuron with linear activation
model.compile(loss='mse', optimizer='adam')

model.fit(x_train, y_train, epochs=2000, batch_size=32, verbose=2)

train_time = time.time() - start_time
train_time
```

```python
model_name = 'model_1_LinearRegression_DNN'
```

```python
model.save(model_name)
```

```python
model = load_model(model_name)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean squared error (MSE): {mse:.2f}")
```

```python
def predict_new_data(new_data):
    new_data = np.reshape(new_data, (1, -1))  # Reshape to fit model input format
    start=time.time()
    prediction = model.predict(new_data)[0][0]  # Extract single prediction value
    end = time.time()
    print(end)
    return prediction

new_features = np.array([6.0, 3.0, 7.0, 0.0, 18.0, 150.0, 396.9, 4.98, 8.0, 9.0, 1.1, 12.2, 13.2])  # Example features (adjust as needed)
predicted_price = predict_new_data(new_features)
print(f"Predicted price for the new data point: ${predicted_price:.2f}")

# print(f"Training time: {train_time:.2f} seconds")
```
